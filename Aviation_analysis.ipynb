Handling the inconsistent time

```python
def correct_times(row):
    # Compare times and adjust ArrivalDate if needed
    if row['ArrivalDate'] == row['DepartureDate'] and row['ArrivalTime'] < row['DepartureTime']:
        # Convert ArrivalDate to datetime, add a day, and convert back to string
        row['ArrivalDate'] = (datetime.strptime(row['ArrivalDate'], '%d/%m/%Y') + timedelta(days=1)).strftime('%d/%m/%Y')
    return row

# Apply the function to correct time issues
df = df.apply(correct_times, axis=1)

# Remove seconds from DepartureTime and ArrivalTime
df['DepartureTime'] = pd.to_datetime(df['DepartureTime'], format='%H:%M:%S').dt.strftime('%H:%M %p')
df['ArrivalTime'] = pd.to_datetime(df['ArrivalTime'], format='%H:%M:%S').dt.strftime('%H:%M %p')

# Convert dates to the desired format (DD/MM/YYYY)
df['DepartureDate'] = pd.to_datetime(df['DepartureDate'], format='%d/%m/%Y').dt.strftime('%d/%m/%Y')
df['ArrivalDate'] = pd.to_datetime(df['ArrivalDate'], format='%d/%m/%Y').dt.strftime('%d/%m/%Y')

print(df)
```

*Creating a Flight duration column*

```python
# Function to correct time format
def correct_time_format(time_str):
    # Check if the input is a string
    if isinstance(time_str, str):
        if 'PM' in time_str and int(time_str.split(':')[0]) < 12:
            # Convert '01:00 PM' to '13:00' etc.
            hour, minute = time_str.split(':')
            return f"{int(hour) + 12}:{minute[:-3]}"  # remove ' PM' and convert hour
        elif 'AM' in time_str and time_str.startswith('12'):
            # Convert '12:00 AM' to '00:00'
            return f"00:{time_str.split(':')[1][:-3]}"  # remove ' AM'
        return time_str[:-3]  # Remove ' AM' or ' PM'
    return time_str  # In case it's not a string, return it as is

# Apply the function to correct DepartureTime and ArrivalTime
df['DepartureTime'] = df['DepartureTime'].apply(correct_time_format)
df['ArrivalTime'] = df['ArrivalTime'].apply(correct_time_format)

# Convert times to datetime for calculation of duration
df['DepartureTime'] = pd.to_datetime(df['DepartureTime'], format='%H:%M', errors='coerce')
df['ArrivalTime'] = pd.to_datetime(df['ArrivalTime'], format='%H:%M', errors='coerce')

# Combine dates and times for calculating duration
df['DepartureDateTime'] = pd.to_datetime(df['DepartureDate'] + ' ' + df['DepartureTime'].dt.strftime('%H:%M'))
df['ArrivalDateTime'] = pd.to_datetime(df['ArrivalDate'] + ' ' + df['ArrivalTime'].dt.strftime('%H:%M'))

# Calculate flight duration
df['FlightDuration'] = (df['ArrivalDateTime'] - df['DepartureDateTime']).dt.total_seconds() / 60  # in minutes

# Print the final DataFrame with FlightNumber and FlightDuration
print(df[['FlightNumber', 'FlightDuration']])
```

```python
import seaborn as sns
def time_to_minutes(time_str):
    if isinstance(time_str, str):
        hour, minute = map(int, time_str[:-3].split(':'))
        if 'PM' in time_str and hour != 12:
            hour += 12
        elif 'AM' in time_str and hour == 12:
            hour = 0
        return hour * 60 + minute
    return None

# Create new columns for DepartureMinutes and ArrivalMinutes
df['DepartureMinutes'] = df['DepartureTime'].apply(time_to_minutes)
df['ArrivalMinutes'] = df['ArrivalTime'].apply(time_to_minutes)

# Convert DelayMinutes to total minutes (from timedelta)
df['DelayMinute'] = df['DelayMinutes'].dt.total_seconds() / 60  # Convert timedelta to minutes
df['DelayMinute'] = df['DelayMinute'].astype(int)

# Create time category based on DepartureMinutes
def time_category(minutes):
    if minutes < 720:  # Before 12 PM
        return 'Morning'
    elif minutes < 1080:  # Before 6 PM
        return 'Afternoon'
    else:
        return 'Evening'

df['TimeCategory'] = df['DepartureMinutes'].apply(time_category)

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a figure
plt.figure(figsize=(14, 6))

# Box Plot
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(x='TimeCategory', y='DelayMinute', data=df, palette='Set2')
plt.title('Delay Minutes by Time Category')
plt.xlabel('Time of Day')
plt.ylabel('Delay Minutes')

# Heatmap
heatmap_data = df.pivot_table(values='DelayMinute', index='DepartureMinutes', columns='TimeCategory', aggfunc='mean')
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.1f', cbar_kws={'label': 'Average Delay (minutes)'})
plt.title('Heatmap of Average Delay Minutes')
plt.xlabel('Time Category')
plt.ylabel('Departure Minutes')

plt.title('Delay Distribution by Departure time and time category')
# Show plot
plt.tight_layout()
plt.show()
```

 Identifying any relationships between flight delays and departure times (e.g., are flights departing later
in the day more likely to be delayed?).

```python



# Calculate the correlation matrix for relevant columns
correlation_matrix = df[['DepartureMinutes', 'DelayMinute']].corr()

# Set up the figure size for better readability
plt.figure(figsize=(8, 6))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Set plot title
plt.title('Correlation between Departure Time and Delay Minutes')

# Show the plot
plt.show()

```

### Diagonal (self-correlation)
The value of `1` along the diagonal (`DepartureMinutes` to `DepartureMinutes` and `DelayMinute` to `DelayMinute`) simply indicates that each variable is perfectly correlated with itself.

### Off-diagonal correlation (0.31)
The correlation coefficient between `DepartureMinutes` and `DelayMinute` is **0.31**. This means there is a **moderate positive correlation** between the time of departure and the delay duration.

- **Positive correlation**: As the `DepartureMinutes` increases (i.e., as flights depart later in the day), the `DelayMinute` also tends to increase.
- However, the correlation is not very strong, meaning that while there is some relationship between later flights and delays, it is not a very strong predictor.

### Conclusion
Flights departing later in the day might experience slightly more delays compared to earlier flights, but this relationship is moderate.


Determining if there is a significant difference in delays between different airlines uing annova test

```python
import pandas as pd
import scipy.stats as stats

# Replace 'Airline' with the actual name of your airline column if different
grouped_data = df.groupby('Airline')['DelayMinute'].apply(list)

# Perform ANOVA
anova_results = stats.f_oneway(*grouped_data)

# Print results
print(f'F-statistic: {anova_results.statistic}, p-value: {anova_results.pvalue}')

# Interpretation
if anova_results.pvalue <= 0.05:
    print("There is a significant difference in delays between different airlines.")
else:
    print("There is no significant difference in delays between different airlines.")
```

Plot showing Average Flight Delay by Airline and Time of Day

```python
def time_category(dep_time):
    hour = int(dep_time.split(':')[0])
    minute = int(dep_time.split(':')[1][:2])
    total_minutes = hour * 60 + minute
    if total_minutes < 720:  # Before 12 PM
        return 'Morning'
    elif total_minutes < 1080:  # Before 6 PM
        return 'Afternoon'
    else:
        return 'Evening'

# Apply the time_category function to create a new column
df['TimeCategory'] = df['DepartureTime'].apply(time_category)

# Calculate average delay per airline and time category
average_delay = df.groupby(['Airline', 'TimeCategory'])['DelayMinute'].mean().reset_index()

# Create a pivot table for stacking
pivot_table = average_delay.pivot(index='Airline', columns='TimeCategory', values='DelayMinute').fillna(0)

# Ensure columns are in the right order
pivot_table = pivot_table[['Morning', 'Afternoon', 'Evening']]

# Set color palette
plt.style.use('seaborn')
colors = sns.color_palette('viridis', 3)

# Plot stacked bar
fig, ax = plt.subplots(figsize=(12, 7))
pivot_table.plot(kind='bar', stacked=True, color=colors, ax=ax, width=0.7)

# Customize the plot
plt.title('Average Flight Delay by Airline and Time of Day', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Airline', fontsize=14, labelpad=10)
plt.ylabel('Average Delay (Minutes)', fontsize=14, labelpad=10)
plt.xticks(rotation=0, fontsize=12)
plt.yticks(fontsize=12)

# Add gridlines
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Customize legend
plt.legend(title='Time of Day', title_fontsize='13', fontsize='12', loc='upper right', bbox_to_anchor=(1.15, 1))

# Add value annotations on bars
for i in range(len(pivot_table)):
    cumulative_sum = 0
    for j in range(len(pivot_table.columns)):
        value = pivot_table.iloc[i][j]
        if value > 0:
            cumulative_sum += value / 2
            ax.text(i, cumulative_sum, f'{value:.1f}', ha='center', va='center', 
                    color='white', fontweight='bold', fontsize=11)
            cumulative_sum += value / 2

# Adjust layout and display
plt.tight_layout()
plt.show()
```
